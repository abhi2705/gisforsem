```python
#import in Libraries
import osgeo.gdal
import struct, numpy, pylab

#create a list to dump in the matrix for each band 
matrixForAllBands = [0] * 256 #sampleImageBands  

def loop_input_plus(enviFileLocation, pixelX, pixelY, pixelWidth, pixelHeight, sampleImageBands=256):
    
    #Define File location 
    #example enviFileLocation = '/home/jon/test2'
    imageDataset = osgeo.gdal.Open(enviFileLocation)
    
    def enviMatrix(pixelX, 
                    pixelY,
                    pixelWidth, 
                    pixelHeight, 
                    sampleImageBands):
    
        # Extract raw data
        band = imageDataset.GetRasterBand(bandNumber)
        
        #byteStringFunction
        byteString = band.ReadRaster(pixelX, pixelY, pixelWidth, pixelHeight)
        
        # Convert to a matrix
        valueType = {osgeo.gdal.GDT_Byte: 'B', osgeo.gdal.GDT_UInt16: 'H'}[band.DataType]
        
        values = struct.unpack('%d%s' % (pixelWidth * pixelHeight, valueType), byteString)
        
        matrix = numpy.reshape(values, (pixelWidth, pixelHeight))
        
        #output the numbers into a list of matrixs
        matrixForAllBands[bandNumber] = matrix
    
    #now do the looping part
    bandVectorPlus = range(1,sampleImageBands)
    for x in bandVectorPlus:
        bandNumber = x
        sampleEnviMatrix = enviMatrix(pixelX,pixelY,pixelWidth, pixelHeight, sampleImageBands)
```

```python
import numpy as np

#build large matrix for all bands
allTheRows = np.zeros( [256,2500] )
longRow = np.zeros( [2500] )


#for loop to step through each of the energy bands matrix
for n in range(1,256):
    sequenceKEVbands = matrixForAllBands[n]
    #startingSpecificRow = sequenceKEVbands[0,:]

    #step through each row in the matrix and merge them together into one large array
    for m in range(0,50):
        #print (n,m)
        
        #pull out each row from the specific keV array
        shortRow = np.array(sequenceKEVbands[m,:])
        #print shortRow
        
        #combine the 50 shortrows into one large row - 1 row for each keV matrix
            #how does it know to switch to a new row when m changes?
            #also how to move over the slots for the matrix?
        for everyFifty in range(0,2500,50):
            longRow[everyFifty] = shortRow 
        
    #put each longRow in the correct Row of the giant array   
    allTheRows[n,:] = longRow
        
#print allTheRows
```
